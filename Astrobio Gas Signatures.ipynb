```python
with open('/Users/Jackroberson/Downloads/Prox_Cent_VPL', 'r') as f:
    text = [line.split() for line in f]

df = pd.DataFrame(text, columns = ["Wavelength", "Wave Number", "Solar Flux Incidence", "Planetary Flux", "Albedo"])

df = df.astype(float)
df.head()
```

```python
df.plot(x = 'Wavelength', y = 'Albedo', figsize = (20, 8), xlim =(0.6, 1.5), ylim = (0, 0.2), title = 'Proxima Centauri b Flux vs. Geometric Albedo')
CH4_lines = [0.79, 0.89, 1.00, 1.15, 1.38, 1.70, 2.310, 3.38]
CO2_lines = [1.050, 1.21, 1.6, 2.010, 2.73]

for xc in CH4_lines:
    plt.axvline(x = xc, color = 'r')

for xc in CO2_lines:
    plt.axvline(x = xc, color = 'g')


plt.legend()
plt.show()
```
